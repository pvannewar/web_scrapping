```python
def has_one_digit_before_and_after(value):
    if isinstance(value, float):  # Ensure the value is a float
        parts = str(value).split(".")  # Split into integer and decimal parts
        return len(parts[0]) == 1 and len(parts[1]) == 1  # Check lengths
    return False

# Filter rows where 'rating' has one digit before and one digit after the decimal
data_filtered = without_duplicates[without_duplicates['Rating'].apply(has_one_digit_before_and_after)]
```
